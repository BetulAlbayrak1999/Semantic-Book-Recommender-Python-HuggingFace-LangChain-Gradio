```python
import time

import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
books= pd.read_csv(f"{path}/books.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax= plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
```

```python
import numpy as np
from datetime import date

books["missing_description"]= np.where(books["description"].isna(), 1, 0)
books["age_of_book"]= date.today().year - books["published_year"]
```

```python
columns_of_interest= ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix= books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(10,10))
heatmap= sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap= "coolwarm",cbar_kws={"label": "Spearman Correlation"})

heatmap.set_title("Correlation Matrix")
plt.show()
```

```python
book_no_missing= books[~books["description"].isna() &
      ~books["num_pages"].isna() &
      ~books["average_rating"].isna()&
      ~books["published_year"].isna()
]
```

```python
book_no_missing
```

```python
book_no_missing["categories"].value_counts().reset_index().sort_values(by="count", ascending=False)
```

```python
book_no_missing.loc[:, "num_words_in_desc"]= book_no_missing["description"].str.split().str.len()
```

```python
book_no_missing
```

```python
book_no_missing.loc[book_no_missing["num_words_in_desc"].between(5, 10), "description"]
```

```python
book_no_missing_above_25_words_desc= book_no_missing[book_no_missing["num_words_in_desc"]>= 25]
```

```python
book_no_missing_above_25_words_desc
```

```python
book_no_missing_above_25_words_desc.loc[:,"title_and_subtitle"]= (np.where(book_no_missing_above_25_words_desc["subtitle"].isna(), book_no_missing_above_25_words_desc["title"], book_no_missing_above_25_words_desc[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)))
```

```python
book_no_missing_above_25_words_desc
```

```python
book_no_missing_above_25_words_desc.loc[:, "tagged_description"]= book_no_missing_above_25_words_desc[["isbn13", "description"]].astype(str).agg(" ".join, axis= 1)
```

```python
book_no_missing_above_25_words_desc
```

```python
(book_no_missing_above_25_words_desc
 .drop(["subtitle", "missing_description", "num_words_in_desc", "age_of_book"], axis= 1)
 .to_csv("books_cleaned.csv", index= False)
 )
```
